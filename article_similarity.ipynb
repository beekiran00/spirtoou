```python
from urllib import response


def fetchAPIPosTone():

    """
    Fetch all the APIs that work with articles ->
    1. volinfo x
    2. volraw x
    3. artlist 
    4. combininng it with tone <0 or tone>0 as params

    """

    userQuery = "Liz Truss"


    
    response = requests.get("https://api.gdeltproject.org/api/v2/doc/doc?query="+userQuery+"&mode=artlist&maxrecords=250&tone>0&format=csv&sourcelang:english")
    filename = userQuery+"_PosToneArt.csv"
    f = open(filename,'w')
    f.write(response.text)
    f.close()


    

    return filename

def fetchAPINegTone():

    """
    Fetch all the APIs that work with articles ->
    1. volinfo x
    2. volraw x
    3. artlist 
    4. combininng it with tone <0 or tone>0 as params

    """

    userQuery = "Truss Resigned"
    

    response = requests.get("https://api.gdeltproject.org/api/v2/doc/doc?query="+userQuery+"&mode=artlist&maxrecords=250&tone<0&format=csv&sourcelang:english")
    filename = userQuery+"_NegToneArt.csv"
    f = open(filename,'w')
    f.write(response.text)
    f.close()

    

    return filename


```

```python
pos = fetchAPIPosTone()
neg = fetchAPINegTone()
```

```python
# Make the +ve and -ve files to run tests on them 
# EXAMPLE TESTS ->
# 1. join the tables to see how they join -> they might join with common elements and 500 rows become less that means, there is some common things. which shouldnt be because the tones are different
# 2. use pandas functions to check common elements, 
# 3. Same link means same title -> perform string matching on the titles

dfPos = pd.read_csv(pos)
dfNeg = pd.read_csv(neg)


```
